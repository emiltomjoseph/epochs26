Movies vs TV Shows

```python
df["type"].value_counts()
```

Graph

```python
df["type"].value_counts().plot(kind="bar")

plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

Top 10 Countries

```python
df["country"].value_counts().head(10)
```

Top Ratings

```python
df["rating"].value_counts().plot(kind="bar")
plt.title("Content Ratings")
plt.show()
```

Top Genres

```python
df["listed_in"].value_counts().head(10).plot(kind="barh")

plt.title("Top Genres")

plt.show()
```

Release Years

```python
df["release_year"].value_counts().sort_index().plot(figsize=(12,5))

plt.title("Content Released by Year")

plt.show()
```

Dataset Overview:
This dataset contains information about Netflix Movies and TV Shows including title, director, cast, country, release year, rating, duration and genres.

Business Problem: How can Netflix improve its recommendation system and understand user preferences using historical content data?

Machine Learning Problem: Recommendation System

Justification: A recommendation system can suggest relevant movies and TV shows based on user preferences and content similarity.

Target Variable: Type (Movie / TV Show)

Key Features:

Country

Release Year

Rating

Duration

Genre

Director

Observations:
1. Movies are more numerous than TV Shows.

2. Netflix content increased rapidly after 2015.

3. Several columns contain missing values such as Director, Country and Cast.
